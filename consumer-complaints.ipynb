```python
avg_complaints = []
for day in range(7):
    x = weekdays_report[weekdays_report['Weekdays'] == day]['Date received'].value_counts().mean()
    avg_complaints.append(x)
    
avg_complaints
```

```python
days = ["Mondays", "Tuesdays", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

plt.plot(avg_complaints)
plt.xticks(range(7), days)
plt.show
```
